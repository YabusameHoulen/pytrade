```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
from scipy.stats import chi2

plt.style.use(["science", "notebook", "grid"])
```

```python
test_size = 4
vals = chi2.ppf(0.683, test_size)  ### 百分点函数，cdf 的倒数
vals/2
```

```python
chi2.cdf(vals, 4)
```

```python
r = chi2.rvs(test_size, size=1000)
```

```python
x = np.linspace(chi2.ppf(0.01, test_size), chi2.ppf(0.99, test_size), 100)
fig, ax = plt.subplots(1, 1)
ax.plot(x, chi2.pdf(x, test_size), "r-", lw=5, alpha=0.6, label="chi2 pdf")
ax.hist(r, density=True, bins="auto", histtype="stepfilled", alpha=0.2)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc="best", frameon=False)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(0, 50, 1000)
ax.set_xlim([0, x[-1]])
ax.set_ylim(bottom=0, top=0.30)
for n, c in zip([4, 8, 10, 14, 20], ["teal", "r", "b", "g", "y"]):
    ax.plot(x, chi2.pdf(x, n), c, lw=2, alpha=0.6, label=f"chi2 {n} pdf")
ax.legend(loc="best", frameon=False)
plt.show()
```
